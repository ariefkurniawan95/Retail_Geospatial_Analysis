#### Libraries Dataset

```python
#import libraries 
import pandas as pd
import numpy as np
import folium
```

#### Load Dataset

```python
#Population density dataset
df = pd.read_excel('pddk_jkt_2019.xlsx')

#Stores Dataset from Google Places API
df_stores = pd.read_excel('stores_df2.xlsx')
```

#### Data Cleaning for the density and stores dataset

```python
df.head()
```

```python
df_stores.columns
```

```python
def store_name_replacer(input):
    if "Alfamart" in input or "alfamart" in input:
        return "Alfamart"
    else:
        return "Indomaret"
```

```python
df_stores['str_name'] = df_stores['store_name'].apply(store_name_replacer)
```

```python
df_stores
```

```python
df_stores['str_name'].value_counts()
```

```python
df_stores = df_stores.drop(columns=['store_name', 'duplicator','Unnamed: 0'])
```

```python
df_stores
```

```python
df_stores_alfamart = df_stores[df_stores['str_name']=='Alfamart']
df_stores_alfamart = df_stores_alfamart.reset_index(drop=True)
df_stores_alfamart
```

```python
df_stores_indomaret = df_stores[df_stores['str_name']=='Indomaret']
df_stores_indomaret = df_stores_indomaret.reset_index(drop=True)
df_stores_indomaret
```

```python
df['nama_kelurahan'] = df['nama_kelurahan'].replace('WIJAYA KUSUMA', 'WIJAYA KESUMA')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('JATIPULO', 'JATI PULO')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('SETIA BUDI', 'SETIABUDI')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('HALIM PERDANA KUSUMAH', 'HALIM PERDANA KUSUMA')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('PULO GADUNG', 'PULOGADUNG')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('PALMERIAM', 'PAL MERIAM')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('PALMERIAM', 'PAL MERIAM')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('JATI PADANG', 'JATIPADANG')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('BALEKAMBANG', 'BALE KAMBANG')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('PONDOK RANGGON', "PONDOK RANGON")
df['nama_kelurahan'] = df['nama_kelurahan'].replace('KALI BARU', 'KALIBARU')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('KERENDANG', 'KRENDANG')
df['nama_kelurahan'] = df['nama_kelurahan'].replace('RAWA JATI', 'RAWAJATI')
```

```python
bin_density = df['kepadatan_jiwa_per_km'].quantile([0,0.25,0.5,0.75,1])
bin_density
```

```python
list_lat_alfamart = df_stores_alfamart['store_lat']
list_long_alfamart = df_stores_alfamart['store_long']
list_add_alfamart = df_stores_alfamart['store_address']

list_lat_indomaret = df_stores_indomaret['store_lat']
list_long_indomaret = df_stores_indomaret['store_long']
list_add_indomaret = df_stores_indomaret['store_address']
```

```python
list_lat_alfamart
```

```python
jakarta_map = folium.Map(
    location = [-6.208518524851216, 106.8444022635577],
    zoom_start = 12)

jakarta_geojson = r'jakarta_by_kelurahan_geo.geojson'

jkt_choro = folium.Choropleth(
    geo_data=jakarta_geojson,     #file geojson (data geospatial)
    data  = df,         #dataframe pandas
    columns = ['nama_kelurahan','kepadatan_jiwa_per_km'],  #dataframe column used
    key_on= 'feature.properties.name',  #for indexing purpose)
    bins = bin_density,
    legend_name = 'Population Density per km square')
    
jkt_choro.add_to(jakarta_map)
jakarta_map
```

```python
jakarta_map = folium.Map(
    location = [-6.208518524851216, 106.8444022635577],
    zoom_start = 12)

jakarta_geojson = r'jakarta_by_kelurahan_geo.geojson'

jkt_choro = folium.Choropleth(
    geo_data=jakarta_geojson,     #file geojson (data geospatial)
    data  = df,         #dataframe pandas
    columns = ['nama_kelurahan','kepadatan_jiwa_per_km'],  #dataframe column used
    key_on= 'feature.properties.name',  #for indexing purpose)
    bins = bin_density,
    legend_name = 'Population Density per km square')

for i in range(len(list_long_alfamart)):
    alfa_pin = folium.features.CustomIcon('alfamart.png', icon_size=(20,12))
    alfamart_pin = folium.Marker(location=[list_lat_alfamart[i], list_long_alfamart[i]],
                            tooltip='Alfamart',
                            popup='',
                            icon = alfa_pin)
    jakarta_map.add_child(alfamart_pin)    
#circle marker
    
for i in range(len(list_long_indomaret)):
    indo_pin = folium.features.CustomIcon('indomaret.png', icon_size=(20,12))
    indomaret_pin = folium.Marker(location=[list_lat_indomaret[i], list_long_indomaret[i]],
                            tooltip='Indomaret',
                            popup='',
                            icon = indo_pin)
    jakarta_map.add_child(indomaret_pin)    
    
jkt_choro.add_to(jakarta_map)
jakarta_map
```
