```python
def location_definer(input):
    if input == "JAKARTA BARAT":
        return ('-6.159958652730745, 106.7487602366644')
    elif input == "JAKARTA UTARA":
        return ('-6.1478435857932965, 106.89347945711535')
    elif input =="JAKARTA PUSAT":
        return ('-6.184630187414843, 106.84518846652338')
    elif input == 'JAKARTA SELATAN':
        return ('-6.274932483754558, 106.82737116971047')
    elif input == 'JAKARTA TIMUR':
        return ('-6.219574258590256, 106.90228209296544')
    else :
        return ('-6.184630187414843, 106.84518846652338')
```

```python
df['latlong'] = df['nama_kabupaten/kota'].apply(location_definer)
```

```python
stores_list=list(['Alfamart','Indomaret'])
kecamatan_list = list(df['nama_kecamatan'])
location_list = list(df['latlong'])
```

```python
print(location_list)
```

```python
querylist = []
querylocation = []
for i in range(len(stores_list)):
    for j in range(len(kecamatan_list)):
        insert_val = (stores_list[i],"+",kecamatan_list[j])
        querylist.append(''.join(insert_val))
        querylocation.append(location_list[j])
```

```python
stores = [] 
for i in range(len(querylist)):
    params = {}
    key = 'insert_your_api_here'
    type = 'convenience_store'
    radius = '7500'
    region = 'id'
    endpoint_url = f"https://maps.googleapis.com/maps/api/place/textsearch/json?query={querylist[i]}&location={querylocation[i]}&radius={radius}&region={region}&type={type}&key={key}"
    res = requests.get(endpoint_url, params=params)
    results = json.loads(res.content)
    stores.extend(results['results'])
    time.sleep(2)
    while "next_page_token" in results:
        params['pagetoken'] = results['next_page_token'],
        res = requests.get(endpoint_url, params = params)
        results = json.loads(res.content)
        stores.extend(results['results'])
        time.sleep(2)
```

```python
store_name = []
store_address = []
store_lat = []
store_long = []

for i in range(len(stores)):
    store = stores[i]
    try:
        store_name.append(store['name'])
    except:
        store_name.append('none')
    try:
        store_address.append(store['formatted_address'])
    except:
        store_address.append('none')
    try:
        store_lat.append(store['geometry']['location']['lat'])
    except:
        store_lat.append('none')
    try:
        store_long.append(store['geometry']['location']['lng'])
    except:
        store_long.append('none')

df_dict = {'store_name': store_name, 'store_address': store_address, 'store_lat': store_lat, 'store_long': store_long}
stores_df = pd.DataFrame(df_dict)
stores_df['duplicator'] = stores_df['store_name'] + stores_df['store_address']
stores_df.drop_duplicates(['duplicator'], inplace=True)
```

```python
stores_df.to_excel('stores_df2.xlsx')
```
